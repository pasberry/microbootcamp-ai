```python
# Import dependencies
import pandas as pd
```

```python
# Read in the CSV file as a Pandas DataFrame
service_ratings_df = pd.read_csv("https://static.bc-edx.com/mbc/ai/m2/datasets/service-ratings.csv")

# Review the DataFrame
service_ratings_df.head()
```

```python
# Visualize a scatter plot of the data
service_ratings_df.plot.scatter(x="mobile_app_rating", y="personal_banker_rating")
```

## Run the k-means model with 2 clusters

```python
# Start by importing the K-means algorithm
from sklearn.cluster import KMeans
```

```python
# Create and initialize the K-means model instance for 2 clusters
model = KMeans(n_clusters=2, n_init='auto',random_state=1)

# Print the model
model
```

```python
# Fit the data to the instance of the model
model.fit(service_ratings_df)
```

```python
# Make predictions about the data clusters using the trained model
customer_ratings = model.predict(service_ratings_df)

# Print the predictions
print(customer_ratings)
```

```python
# Create a copy of the DataFrame
service_rating_predictions_df = service_ratings_df.copy()

# Add a column to the DataFrame that contains the customer_ratings information
service_rating_predictions_df['customer rating'] = customer_ratings

# Review the DataFrame
service_rating_predictions_df.head()
```

```python
# Plot the data points based on the customer rating
service_rating_predictions_df.plot.scatter(
    x="mobile_app_rating",
    y="personal_banker_rating",
    c="customer rating",
    colormap='winter')
```

```python
# Create and initialize the K-means model instance for 3 clusters
model2 = KMeans(n_clusters=3, n_init='auto',random_state=1)

# Print the model
```

```python
model2.fit(service_ratings_df)
customer_ratings2 = model2.predict(service_ratings_df)

service_rating_predictions_df2 = service_ratings_df.copy()
service_rating_predictions_df2['customer rating'] = customer_ratings2

service_rating_predictions_df2.head()
```

```python
service_rating_predictions_df2.plot.scatter(
    x="mobile_app_rating",
    y="personal_banker_rating",
    c="customer rating",
    colormap='winter')
```

```python
# Create and initialize the K-means model instance for 3 clusters
model3 = KMeans(n_clusters=4, n_init='auto',random_state=1)

model3.fit(service_ratings_df)
customer_ratings3 = model3.predict(service_ratings_df)

service_rating_predictions_df3 = service_ratings_df.copy()
service_rating_predictions_df3['customer rating'] = customer_ratings3

service_rating_predictions_df3.head()

service_rating_predictions_df3.plot.scatter(
    x="mobile_app_rating",
    y="personal_banker_rating",
    c="customer rating",
    colormap='winter')
```
