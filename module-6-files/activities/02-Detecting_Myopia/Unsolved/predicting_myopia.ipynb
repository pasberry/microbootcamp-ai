# Predicting Myopia

In this activity, you'll use a deep learning model to predict whether a person has myopia.

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import tensorflow as tf
```

```python
# Import our input dataset
file_path = "https://static.bc-edx.com/mbc/ai/m6/datasets/myopia.csv"

df = pd.read_csv(file_path)

# Display sample data
df
```

```python
from sklearn.model_selection import train_test_split

# Create the features and target sets
X = df.copy().drop('MYOPIC', axis=1)
y = df['MYOPIC']
```

```python
# Preprocess numerical data for neural network

# Create a StandardScaler instances
X_scaler = StandardScaler()

# Scale the data
X_scaler.fit(X)

# Scale the features data
X_train_scaled = X_scaler.transform(X)
```

```python
# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X_train_scaled, y, random_state=1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the deep learning model 
number_input_features = 14
hidden_nodes_layer1 = 8
hidden_nodes_layer2 = 4

nn = Sequential()

# Add the first hidden layer
nn.add(Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Add the second hidden layer
nn.add(Dense(units=hidden_nodes_layer2, activation="relu"))

# Add the output layer
nn.add(Dense(units=1, activation="sigmoid"))
```

```python
# Compile the Sequential model together and customize metrics
nn.compile(loss="mean_squared_error", optimizer="adam", metrics=["mse"])
```

```python
# Train the model
deep_net_model = nn.fit(X_train, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model1_loss, model1_mse = nn.evaluate(X_test, y_test, verbose=2)
```

```python
# Make predictions on the test data
predictions = nn.predict(X_test).round().astype("int32")
```

```python
# Create a DataFrame to compare the predictions with the actual values
results = pd.DataFrame({"predictions": predictions.ravel(), "actual": y_test})

# Display sample data
results.head(10)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, results['predictions'])
```
